# Q2:

#### I created a new dataframe that contains only a region that begins with the first letter of my name, and the year 2016. I have shown how many lines there are in my dataframe.

```python
s_data=data[data['region'].str.startswith('S')].copy()
s_data_2016= s_data[(s_data.region=='SanFrancisco') & (s_data.year==2016)].copy()
len(s_data_2016)
```

# Q3 :

#### On the same dataframe I created in question 2: The average sales is:

```python
 s_data_2016.total_sold.mean()
```

# Q4 :

#### On the same dataframe I created in question 2: Total amount of avocado bags ('Total Bags' column) sold in September (month number 9 in 'Date' column) is:


```python
s_data_2016['date'] = pd.to_datetime(s_data_2016.Date)
s_data_2016['month'] = pd.DatetimeIndex(s_data_2016.date).month
s_data_september = s_data_2016[s_data_2016.month == 9]
s_data_september['Total Bags'].sum()
```

# Q5 :

```python
ax=data.pivot_table("total_sold","type",aggfunc=sum).plot(kind="bar",colormap="GnBu_r")
ax.set_title('comparing sales between types of avocado',fontsize=16)
ax.set_ylabel("total sales",fontsize=12)
ax.set_xlabel(" conventiona/organic")
```
