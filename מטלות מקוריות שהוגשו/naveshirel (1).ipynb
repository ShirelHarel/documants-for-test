**Solution to question 9:**
In order to know what is bigger- the mean or the median for the column "daily_vaccinations", we calculate the mean and the median and compered 

```python
median_Norway = vacc_Norway_df['daily_vaccinations'].median()
mean_Norway = vacc_Norway_df['daily_vaccinations'].mean()
if median_Norway > mean_Norway:
    print ('median is bigger than mean')
else:
    print ('mean is bigger than median')
        
```
