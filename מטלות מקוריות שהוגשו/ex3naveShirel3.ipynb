solution to question 3:

We calculated the correlation according to Spearman between numbers of 'likes','dislikes','views','comment_count'

```python

corelation_spearman = yout_df[['likes','dislikes','views','comment_count']].corr(method='spearman')
corelation_spearman
```

solution to question 4:

We created a heatmap to a question 3.

```python
corr = sns.heatmap(data=corelation_spearman , cmap='Greens',annot= True)
corr
```

solution to question 5:
    
We found the strongest correlations from question 3, and we created a scatterplot for both of them.

```python
import matplotlib.pyplot as plt
fig ,axes = plt.subplots(1, 2,figsize = (16,8) )
highest = corelation_spearman.unstack().sort_values()[-8:-4]
highest_corrs = highest.iloc[1:3]
titles= pd.Series.keys(highest_corrs)
f_fig= sns.scatterplot(ax=axes[0], data = yout_df, x = titles[0][0] , y=titles[0][1])
f_fig2= sns.scatterplot(ax=axes[1], data = yout_df, x = titles[1][0] , y=titles[1][1])

```
